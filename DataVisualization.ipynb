```python
datasetpath = '/content/ddrive/MyDrive/matches_1930_2022.csv'
```

```python
ScoredAboveZero = df['Date']>'0-0'
```

```python
df[ScoredAboveZero]
```

df.shape

```python
ScoreByBrazil = df['home_team'] == 'Brazil'
```

```python
df['Score'].value_counts()
```

```python
brazildf=df[ScoredAboveZero &ScoreByBrazil].sort_values(by ='home_team',ascending=True)
```

```python
brazildf
```

```python
brazildf.info()
```

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
```

```python
brazildf.plot(x='home_score',y='away_score',kind='scatter')
plt.xlabel('Home score')
plt.ylabel('Away score')
plt.legend('Take away')
plt.title('Brazil Matches')
plt.show()
```

```python
brazildf.plot(x='home_score',y='away_score',kind='bar')
plt.xlabel('Home score')
plt.ylabel('Away score')
plt.legend('Take away')
plt.title('Brazil Matches')
plt.show()
```

```python
import pandas as pd
```

```python
datasatPath = '/content/drive/MyDrive/retail_transactions.csv'
df = pd.read_csv(datasatPath)
```

```python
df
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
```

```python
retail_df = df
```

```python
print("First 15 rows of the DataFrame:")
print(retail_df.head(15))
```

```python
print("\nDataFrame Info:")
print(retail_df.info())
```

```python
print("\nDataFrame Description:")
print(retail_df.describe())
```

```python
print("\nMissing values per column:")
print(retail_df.isnull().sum())
```

```python
retail_df['Discount'] = retail_df['Discount'].fillna(0)
```

```python
print("\nMissing values after filling Discount:")
print(retail_df['Discount'].isnull().sum())
```

```python
retail_df['NetPrice'] = retail_df['Price'] * (1 - retail_df['Discount'] / 100)
```

```python
retail_df['TotalSale'] = retail_df['Quantity'] * retail_df['NetPrice']
```

```python
print("\nDataFrame with new NetPrice and TotalSale columns:")
print(retail_df.head())
```

```python
filtered_df = retail_df[retail_df['TotalSale'] >= 20].copy()
```

```python
filtered_df
```

```python
print("\nShape of the filtered DataFrame:")
print(filtered_df.shape)
```

```python
filtered_df['PaymentType'] = filtered_df['PaymentType'].astype('category')
```

```python
print("\nData types after converting PaymentType to category:")
print(filtered_df.dtypes)
```

```python
grouped = filtered_df.groupby('ProductCategory').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalSale=('TotalSale', 'sum')
)
```

```python
filtered_df.groupby
```

```python
grouped_sorted = grouped.sort_values(by='TotalSale', ascending=False)
```

```python
grouped_sorted
```

```python
print("\nGrouped and sorted by TotalSale:")
print(grouped_sorted)
```

```python
top_3_categories = grouped_sorted.head(3)
print("\nTop 3 product categories with the highest total sales:")
print(top_3_categories)
```

```python
multi_grouped = filtered_df.groupby(['StoreID', 'ProductCategory'])['TotalSale'].mean().reset_index()
```

```python
multi_grouped
```

```python
multi_grouped = multi_grouped.rename(columns={'TotalSale': 'AvgTotalSale'})
```

```python
print("\nMulti-level grouping by StoreID and ProductCategory with average TotalSale:")
print(multi_grouped)
```

```python
multi_grouped = filtered_df.groupby(['StoreID', 'ProductCategory'])['TotalSale'].mean().reset_index()
multi_grouped = multi_grouped.rename(columns={'TotalSale': 'AvgTotalSale'})
sorted_grouped = multi_grouped.sort_values(['StoreID','AvgTotalSale'], ascending=[True, False])
```

```python
sorted_grouped
```

```python
top_categories_per_store = sorted_grouped.groupby('StoreID').first().reset_index()
```

```python
summary_df = top_categories_per_store[['StoreID', 'ProductCategory', 'AvgTotalSale']]
```

```python
summary_df
```

```python
print("\nSummary DataFrame with top ProductCategory and average TotalSale for each store:")
print(summary_df)
```

```python
plt.scatter(filtered_df['Quantity'], filtered_df['NetPrice'], color='red', alpha=0.5)
plt.xlabel('Quantity')
plt.ylabel('Net Price')
plt.title('Quantity vs Net Price')
plt.show()
```

```python
plt.plot(grouped.index, grouped['TotalQuantity'], marker='o', linestyle='-', color='green')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.title('Total Quantity Sold by Product Category')
plt.show()
```

```python
multi = filtered_df.groupby(["StoreID", 'ProductCategory'])['TotalSale'].mean().reset_index()

multi
```

```python
multi_grouped = multi.rename(columns={'TotalSale': 'AvgTotalSale'})  # Assign the result of rename to multi_grouped
```

```python
print("\nMulti-level grouping by StoreID and Product Category with average TotalSale:")

print(multi_grouped)
```

```python
multi_grouped = filtered_df.groupby(['StoreID', 'ProductCategory'])['TotalSale'].mean().reset_index()
multi_grouped = multi_grouped.rename(columns={'TotalSale': 'AvgTotalSale'})
sorted_grouped = multi_grouped.sort_values(['StoreID', 'AvgTotalSale'], ascending=[True, False])
```

```python
top_categories_per_store=sorted_grouped.groupby('StoreID').first().reset_index()
```

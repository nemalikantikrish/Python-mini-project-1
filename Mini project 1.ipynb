```python
class ToDoList:
    def __init__(self):
        # Initialize an empty list to hold tasks
        self.tasks = []

    def create_task(self, task):
        # Add a new task to the list
        self.tasks.append(task)
        print(f'Task "{task}" created.')

    def see_tasks(self):
        # Display all tasks in the list
        if not self.tasks:
            print("Your to-do list is empty.")
        else:
            print("Your to-do list:")
            # Loop through the tasks without using enumerate
            for i in range(len(self.tasks)):
                print(f"{i + 1}. {self.tasks[i]}")  # Display task number and task

    def update_task(self, task_number, new_task):
        # Update an existing task at the specified index
        if 0 < task_number <= len(self.tasks):
            old_task = self.tasks[task_number - 1]  # Get the old task
            self.tasks[task_number - 1] = new_task  # Update the task
            print(f'Task "{old_task}" updated to "{new_task}".')
        else:
            print("Invalid task number.")  # Handle invalid input

    def remove_task(self, task_number):
        # Remove a task from the list based on its number
        if 0 < task_number <= len(self.tasks):
            removed_task = self.tasks.pop(task_number - 1)  # Remove the task
            print(f'Task "{removed_task}" removed from the list.')
        else:
            print("Invalid task number.")  # Handle invalid input

    def delete_all(self):
        # Clear all tasks from the list
        self.tasks.clear()
        print("All tasks deleted from the list.")

def main():
    # Create an instance of the ToDoList class
    todo_list = ToDoList()

    while True:
        # Display the menu options to the user
        print("\nTo-Do List Menu:")
        print("1. Create Task")
        print("2. See Tasks")
        print("3. Update Task")
        print("4. Remove Task")
        print("5. Delete All Tasks")
        print("6. Exit")
        choice = input("Choose an option (1-6): ")

        # Use match-case to handle user input
        match choice:
            case '1':
                # Create a new task
                task = input("Enter the task to create: ")
                todo_list.create_task(task)
            case '2':
                # See all tasks
                todo_list.see_tasks()
            case '3':
                # Update an existing task
                task_number = int(input("Enter the task number to update: "))
                new_task = input("Enter the new task: ")
                todo_list.update_task(task_number, new_task)
            case '4':
                # Remove a task
                task_number = int(input("Enter the task number to remove: "))
                todo_list.remove_task(task_number)
            case '5':
                # Delete all tasks
                todo_list.delete_all()
            case '6':
                # Exit the program
                print("Exiting the program.")
                return  # Exit the function to stop the program
            case _:
                # Handle invalid options
                print("Invalid option. Please choose again.")

if __name__ == "__main__":
    main()  # Start the program
```
